```python
import numpy as np

scores = np.array([45, 67, 89, 32, 90, 76, 50])
print("Original scores:", scores)

# Boolean mask for scores above 60
high_scores = scores[scores > 60]
print("Scores above 60:", high_scores)

# Select every second score
alternate_scores = scores[::2]
print("Every second score:", alternate_scores)
```

```python
import numpy as np

temps = np.array([22.5, 23.0, 21.0, 25.5, 24.0])
print("Temperatures:", temps)

mean_temp = np.mean(temps)
print("Average Temperature:", mean_temp)

std_dev = np.std(temps)
print("Standard Deviation:", std_dev)

temp_diff = temps - mean_temp
print("Deviation from Mean:", temp_diff)
```

```python
import numpy as np

marks = np.array([35, 78, 90, 55, 29, 65, 84])
print("Marks:", marks)

# Identify passed students (marks ≥ 40)
passed = marks >= 40
print("Passed mask:", passed)

# Replace failed marks with 0
marks_cleaned = np.where(passed, marks, 0)
print("Cleaned Marks (failures set to 0):", marks_cleaned)

# Check if all students passed
print("All passed?", np.all(passed))
```

```python
import numpy as np

readings = np.array([1.1, np.nan, 2.4, 3.3, np.nan, 4.8])
print("Original Readings:", readings)

# Find and count NaNs
nan_mask = np.isnan(readings)
print("Missing readings at:", np.where(nan_mask)[0])

# Fill NaNs with mean of non-missing values
mean_val = np.nanmean(readings)
readings[nan_mask] = mean_val

print("Filled Readings:", readings)
```

```python
import numpy as np

np.random.seed(10)  # Reproducibility
dice_rolls = np.random.randint(1, 7, size=20)  # Simulate 20 dice rolls
print("Dice Rolls:", dice_rolls)

# Frequency of each face (1 to 6)
for face in range(1, 7):
    count = np.sum(dice_rolls == face)
    print(f"Face {face} appeared {count} times")

# Average of dice rolls
print("Average roll:", np.mean(dice_rolls))
```
