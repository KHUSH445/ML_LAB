AND

```python
import numpy as np

# Step 1: Define dataset (logical AND function)
# Inputs: 4 samples of 2 features each
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Desired output (AND gate)
Y = np.array([0, 0, 0, 1])

# Step 2: Initialize parameters
learning_rate = 0.1
epochs = 10
weights = np.zeros(X.shape[1])
bias = 0

# Step 3: Hebbian Learning Rule
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    for i in range(len(X)):
        x = X[i]
        y = Y[i]

        # Compute neuron output (dot product)
        output = np.dot(x, weights) + bias

        # Hebbian weight update
        delta_w = learning_rate * x * y
        weights += delta_w
        bias += learning_rate * y

        print(f"Input: {x}, Target: {y}, Δw: {delta_w}, Weights: {weights}, Bias: {bias}")
    print("--------------------------------------------------")

# Step 4: Test final model
print("Final Weights:", weights)
print("Final Bias:", bias)

# Test on all inputs
for x in X:
    y_pred = np.dot(x, weights) + bias
    print(f"Input: {x}, Predicted Output: {y_pred:.2f}")
```
